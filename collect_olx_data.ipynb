```python
import requests
from bs4 import BeautifulSoup as BS
import numpy as np
```

```python
site = "https://www.olx.kz/"
search = "/q-%D0%B0%D1%80%D0%B5%D0%BD%D0%B4%D0%B0/?page="


# All categories for rent ads
general = [ 
              'elektronika',
              'dom-i-sad',
              'rabota',
              'moda-i-stil', 
              'detskiy-mir', 
              'hobbi-otdyh-i-sport',
              'zhivotnye',
              'transport',
             ]

special = {'nedvizhimost': ['kvartiry',
                           'doma',
                           'komnaty',
                           'zemlya',
                           'garazhi-stoyanki',
                           'kommercheskie-pomeshcheniya',
                           'zarubezhnaya-nedvizhimost' ] ,
           'uslugi': ['razvlecheniya',
                      'krasota-i-zdorove',
                      'dlya-biznesa',
                      'perevozki-i-skladskie-uslugi',
                      'remont-i-stroitelstvo', 
                      'uborka',
                      'prokat-tovarov',
                      'avto-uslugi',
                      'bytovye-uslugi',
                      'remont-i-obsluzhivanie-tehniki',
                      'ohrana-bezopasnost',
                      'delovye-uslugi',
                      'foto-i-videosemka',
                      'uslugi-dlya-zhivotnyh',
                      'prochie-uslugi']
           
          }


# OLX displays only 25 pages from category, so this will be our limit
max_pages = 25

# Lists to collect data
gen_titles, gen_prices, gen_locations, gen_published, gen_category = [], [], [], [], []

spec_titles, spec_prices, spec_locations, spec_published, spec_category = [], [], [], [], []




```

```python
# Collecting from general categories
def collect_gen(category):
    
    for page_num in np.arange(1, max_pages+1):
          
        curr_page = requests.get(site + category + search + str(page_num) ) 

        # Open current page with BeatifulSoup
        html = BS(curr_page.content, 'html.parser')

        # Add title of ad
        
        for title in html.find_all('tr', class_='wrap'):
            gen_titles.append(title.strong.text) 
            gen_category.append(category)

        for p in html.find_all('div', class_='space inlblk rel'):
            try:
                price = p.strong.text
                gen_prices.append(price)
            except:
                gen_prices.append('')

        # Add locations
        
                
        for l in html.find_all('td', class_='bottom-cell'): gen_locations.append(l.span.text)



        
        temp = html.find_all('td', class_='bottom-cell')

        for last_div in temp:
            
            content = last_div.getText()
            date_found = content.split()[-2]
            time_found = content.split()[-1]
            
            gen_published.append(time_found + ' ' + date_found)
               
                    
        
```

```python
# Collect special data
def collect_spec(category):                   
    
    for spec in special[category]:
        
        for page_num in np.arange(1, max_pages+1):
          
            curr_page = requests.get(site + category + '/' + spec + search + str(page_num) ) 

            # Open current page with BeatifulSoup
            html = BS(curr_page.content, 'html.parser')

            # Title
            for title in html.find_all('tr', class_='wrap'): 
                spec_titles.append(title.strong.text)
                spec_category.append(category)


            for p in html.find_all('div', class_='space inlblk rel'):
                try:
                    price = p.strong.text
                    spec_prices.append(price)
                except:
                    spec_prices.append('')
                

            
                    
            for l in html.find_all('td', class_='bottom-cell'): spec_locations.append(l.span.text) 



            
            temp = html.find_all('td', class_='bottom-cell')

            for last_div in temp: 
                
                content = last_div.getText()
                date_found = content.split()[-2]
                time_found = content.split()[-1]

                spec_published.append(time_found + ' ' + date_found)
                    
            
            
```

```python

list(map(lambda x: collect_gen(x), general))
    

list(map(lambda x: collect_spec(x), special.keys()))
```

```python
# Join different lists


titles = gen_titles + spec_titles
prices = gen_prices + spec_prices
locations = gen_locations + spec_locations
published = gen_published + spec_published
categories = gen_category + spec_category 
```

```python
import pandas as pd

df = pd.DataFrame({'Title': titles, 'Price': prices, 'Location': locations, 'Published': published, 'Category': categories})
```

```python
df.to_csv('olx_data.csv', index = False)
```
